```python

class Student:
  def __init__(self, name, surname, gender):
    self.name = name
    self.surname = surname
    self.gender = gender
    self.finished_courses = []
    self.courses_in_progress = []
    self.grades = {}
    self.courses_attached = {}


  def __mean_student(self):
    val = 0
    for key, val in list(self.grades.items()):
      if sum(val) == 0:
        return val 
    return f'{key}:{sum(val) / len(val)}'


  def some_student(self, name, surname):
      super().__init__(name, surname)
  def __str__(self):
      return f'Имя: {self.name}\nФамилия: {self.surname}\nСредняя оценка за домашние задания: {self.__mean_student()}\nКурсы в процессе изучения: {", ".join(self.courses_in_progress)}\nЗавершенные курсы: {", ".join(self.finished_courses)}' 
 
  def __eq__(self, other):
     return isinstance(other, Student) and self.__mean_student() == other.__mean_student()

  def __lt__(self, other):
     return isinstance(other, Student) and self.__mean_student() < other.__mean_student()

  def __gt__(self, other):
     return isinstance(other, Student) and self.__mean_student() > other.__mean_student()

  def grade_for_a_lecturer(self, lecturer, course, grade):
    if isinstance(lecturer, Lecturer) and course in self.courses_in_progress and course in lecturer.courses_attached:
      if course in lecturer.grades:
        lecturer.grades[course] += [grade]
      else:
        lecturer.grades[course] = [grade] 
    else: 'Ошибка'    
 
  def student_statistics(students, course):
    grade_val = 0
    grade_num = 0
    for student in students:
        grade_val += sum(student.grades.get(course))
        grade_num += len(student.grades.get(course))
    grade_stats = round(grade_val/grade_num,1)
    return grade_stats
   

class Mentor:
  def __init__(self, name, surname):
    self.name = name
    self.surname = surname   

class Lecturer(Mentor):
  def __init__(self, name, surname):
    super().__init__(name, surname)
    self.courses_attached = []
    self.grades = {}
  
       
  def __mean_lecturer(self):
    val = 0
    for key, val in list(self.grades.items()):
      if sum(val) == 0:
        return val 
    return f'{key}:{sum(val) / len(val)}' 
       

  def some_lecturer(self, name, surname):
      super().__init__(name, surname)
  def __str__(self):
      return f'Имя: {self.name}\nФамилия: {self.surname}\nСредняя оценка за лекции: {self.__mean_lecturer()}' 
  
  def __eq__(self, other):
    return isinstance(other, Lecturer) and self.__mean_lecturer() == other.__mean_lecturer()

  def __lt__(self, other):
    return isinstance(other, Lecturer) and self.__mean_lecturer() < other.__mean_lecturer()   

  def __gt__(self, other):
    return isinstance(other, Lecturer) and self.__mean_lecturer() > other.__mean_lecturer()   

  def lecturer_statistics(lecturer, course):
    grade_val = 0
    grade_num = 0
    for lecturer in lectures:
        grade_val += sum(lecturer.grades.get(course))
        grade_num += len(lecturer.grades.get(course))
    grade_stats = round(grade_val/grade_num,1)
    return grade_stats

class Reviewer(Mentor):
  def __init__(self,name, surname):
    super().__init__(name, surname)
    self.courses_attached = []

  def rate_hw(self, student, course, grade):
    if isinstance(student, Student) and course in self.courses_attached and course in student.courses_in_progress:
      if course in student.grades:
          student.grades[course] += [grade]
      else:
          student.grades[course] = [grade]
    else:
      return 'Ошибка'

  def some_reviewer(self, name, surname):
      super().__init__(name, surname)
  def __str__(self):
      return f'Имя: {self.name}\nФамилия: {self.surname}' 


first_student = Student('Ruoy', 'Eman', 'boy')
first_student.finished_courses += ['Введение в программирование']
first_student.courses_in_progress += ['Python']
first_student.courses_in_progress += ['Git']  
first_reviewer = Reviewer('Some', 'Buddy')
first_reviewer.courses_attached += ['Python']
first_reviewer.rate_hw(first_student, 'Python', 10)
first_reviewer.rate_hw(first_student, 'Python', 10)
second_reviewer = Reviewer('Ted', 'Vuddy')
second_reviewer.courses_attached += ['Python']
second_reviewer.rate_hw(first_student, 'Python', 9)
second_reviewer.rate_hw(first_student, 'Python', 10)
second_student = Student('Bob', 'Grenn', 'girl')
second_student.finished_courses += ['Введение в программирование']
second_student.courses_in_progress += ['Python']
second_student.courses_in_progress += ['Git']
first_reviewer.rate_hw(second_student, 'Python', 10)
first_reviewer.rate_hw(second_student, 'Python', 9)
second_reviewer.rate_hw(second_student, 'Python', 8)
second_reviewer.rate_hw(second_student, 'Python', 10)
first_lecturer = Lecturer('Sam', 'Smith')
second_lecturer = Lecturer('Tom', 'Kruz')
first_lecturer.courses_attached += ['Git']
second_lecturer.courses_attached += ['Git']
first_student.grade_for_a_lecturer(second_lecturer, 'Git', 10)
second_student.grade_for_a_lecturer(second_lecturer, 'Git', 9)
first_student.grade_for_a_lecturer(first_lecturer, 'Git', 10)
second_student.grade_for_a_lecturer(first_lecturer, 'Git', 10)
first_lecturer.grades['Git'] += [10]
second_lecturer.grades['Git'] += [8]
first_student.grades['Python'] += [10]
second_student.grades['Python'] += [10]
students = [first_student, second_student]
lectures = [first_lecturer, second_lecturer]

print(first_reviewer)
print(second_reviewer)
print(first_lecturer)
print(second_lecturer)
print(first_student)
print(second_student)
print(first_lecturer._Lecturer__mean_lecturer() == second_lecturer._Lecturer__mean_lecturer())
print(first_student._Student__mean_student() == second_student._Student__mean_student())
print(first_lecturer._Lecturer__mean_lecturer() < second_lecturer._Lecturer__mean_lecturer())
print(first_student._Student__mean_student() < second_student._Student__mean_student())
print(first_lecturer._Lecturer__mean_lecturer() > second_lecturer._Lecturer__mean_lecturer())
print(first_student._Student__mean_student() > second_student._Student__mean_student())
print('Средняя оценка  за домашние задания по всем студентам в рамках курса Python: ', Student.student_statistics(students, 'Python'))
print('Средняя оценка за лекции всех лекторов в рамках курса Git: ',Lecturer.lecturer_statistics(lectures, 'Git'))
```
